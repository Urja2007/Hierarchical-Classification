```python
import os
import random
import time
import copy
from collections import Counter, defaultdict

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, Subset, Dataset
from torchvision import transforms, datasets, models
from torchvision.transforms import InterpolationMode

from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.metrics import silhouette_score, classification_report, confusion_matrix
```

```python
# Replace transforms for val/test subsets to eval_transforms
# random_split returns Subset objects; we wrap them in a small Dataset to change transforms
class SubsetWithTransform(Dataset):
    def __init__(self, subset, transform):
        self.subset = subset
        self.transform = transform

    def __len__(self):
        return len(self.subset)

    def __getitem__(self, idx):
        img, lbl = self.subset[idx]
        # original dataset has transform baked in; need raw PIL from dataset_all
        # To get raw PIL we access dataset_all.samples? For EuroSAT, easier: re-create dataset with eval transforms referencing indices
        # We'll assume dataset_all has `imgs` / `samples`, but to keep consistent, instead create a wrapper that uses dataset_all with index mapping
        # USE index mapping from subset.indices
        real_idx = self.subset.indices[idx]
        pil_img, lbl = dataset_all.dataset[0] if False else None  # placeholder - will override below

# Simpler approach: re-create datasets using targets and indices
# Build list of all image file paths via dataset_all (it stores samples)
base_eurosat = datasets.EuroSAT(root=DATA_DIR, download=False, transform=None)  # no transform, returns PIL images
all_imgs = base_eurosat.imgs  # list of (path, class_idx)

train_indices = train_ds.indices
val_indices = val_ds.indices
test_indices = test_ds.indices
```

```python
class PathIndexDataset(Dataset):
    def __init__(self, imgs_list, indices, transform):
        # imgs_list: list of (path, target)
        self.selected = [imgs_list[i] for i in indices]
        self.transform = transform

    def __len__(self):
        return len(self.selected)

    def __getitem__(self, idx):
        path, label = self.selected[idx]
        from PIL import Image
        img = Image.open(path).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, label
```

```python
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights

def create_backbone(drop_fc=True, pretrained=True):
    weights = EfficientNet_B0_Weights.IMAGENET1K_V1 if pretrained else None
    backbone = efficientnet_b0(weights=weights)
    if drop_fc:
        # remove classifier, use features
        backbone.classifier = nn.Identity()
        # The feature dimension for EfficientNet-B0 is 1280 after global average pooling
        feat_dim = 1280
    else:
        feat_dim = backbone.classifier[1].in_features  # if not identity
    return backbone, feat_dim

# SupCon model = backbone + projection MLP
class SupConModel(nn.Module):
    def __init__(self, feat_dim=1280, proj_dim=128):
        super().__init__()
        self.backbone, _ = create_backbone(drop_fc=True, pretrained=True)
        # pooling already inside backbone outputs [B, 1280] after adaptive pool and flatten if we use classifier removed.
        # add projection head
        self.projection = nn.Sequential(
            nn.Linear(feat_dim, feat_dim//2),
            nn.ReLU(inplace=True),
            nn.Linear(feat_dim//2, proj_dim)
        )

    def forward(self, x):
        # backbone returns features (batch, feat_dim) if classifier removed
        f = self.backbone(x)
        if isinstance(f, torch.Tensor) and f.ndim == 4:
            # sometimes backbone outputs [B,C,H,W] if classifier removal changed shape - do adaptive pool
            f = nn.AdaptiveAvgPool2d((1,1))(f)
            f = torch.flatten(f, 1)
        z = self.projection(f)
        z = nn.functional.normalize(z, dim=1)
        return z, f
```

```python
class SupConLoss(nn.Module):
    def __init__(self, temperature=0.07):
        super().__init__()
        self.temperature = temperature
        self.eps = 1e-7

    def forward(self, features, labels):
        """
        features: tensor of shape [batch_size, proj_dim] (assumes normalized)
        labels: tensor shape [batch_size]
        Implementation assumes one view per sample. For multi-view, you'd create 2x augmentations per sample.
        We will create positive pairs by matching labels in the batch.
        """
        device = features.device
        batch_size = features.shape[0]
        # cosine similarity matrix
        sim_mat = torch.div(torch.matmul(features, features.T), self.temperature)
        # mask for positives
        labels = labels.contiguous().view(-1, 1)
        mask = torch.eq(labels, labels.T).float().to(device)
        # For numerical stability remove diag
        logits_max, _ = torch.max(sim_mat, dim=1, keepdim=True)
        logits = sim_mat - logits_max.detach()
        # mask out self-contrast cases
        mask_no_self = mask - torch.eye(batch_size, device=device)
        exp_logits = torch.exp(logits) * (1 - torch.eye(batch_size, device=device))
        # denominator = sum_j exp(logits_ij) excluding i
        denom = exp_logits.sum(dim=1, keepdim=True) + self.eps
        # numerator = sum_{positive j} exp(logits_ij)
        numerator = (exp_logits * mask_no_self).sum(dim=1, keepdim=True)
        # loss for each sample
        loss = -torch.log((numerator + self.eps) / denom)
        # only average over samples that have positives (numerator>0)
        valid = (mask_no_self.sum(dim=1) > 0).float()
        loss = (loss.squeeze() * valid).sum() / (valid.sum() + 1e-7)
        return loss
```

```python
# 6) SupCon pretraining loop

def supcon_pretrain(model, train_dataset, val_dataset, epochs=10, lr=1e-4, batch_size=32, out_dir=OUT_DIR):
    model = model.to(DEVICE)
    criterion = SupConLoss(temperature=0.07)
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-5)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=NUM_WORKERS)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=NUM_WORKERS)

    best_loss = 1e9
    history = {"train_loss": [], "val_loss": []}

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        it = 0
        for imgs, labels in tqdm(train_loader, desc=f"SupCon train epoch {epoch+1}/{epochs}", leave=False):
            imgs = imgs.to(DEVICE)
            labels = labels.to(DEVICE)
            z, _ = model(imgs)
            loss = criterion(z, labels)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            it += 1
        train_loss = running_loss / max(1, it)

        # val
        model.eval()
        running_val = 0.0
        it = 0
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs = imgs.to(DEVICE)
                labels = labels.to(DEVICE)
                z, _ = model(imgs)
                loss = criterion(z, labels)
                running_val += loss.item()
                it += 1
        val_loss = running_val / max(1, it)

        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)

        print(f"[SupCon] Epoch {epoch+1}/{epochs} train_loss={train_loss:.4f} val_loss={val_loss:.4f}")

        # checkpoint best
        if val_loss < best_loss:
            best_loss = val_loss
            torch.save(model.state_dict(), os.path.join(out_dir, "supcon_best.pth"))
    return model, history


# Instantiate SupCon model
backbone_for_supcon, feat_dim = create_backbone(drop_fc=True, pretrained=True)
supcon_model = SupConModel(feat_dim=feat_dim, proj_dim=128)
print("SupCon model ready. Feature dim:", feat_dim)

# Pretrain SupCon (this can be slow on CPU)
print("Starting SupCon pretraining... (this may take long on CPU)")
supcon_model, supcon_hist = supcon_pretrain(supcon_model, train_dataset, val_dataset,
                                            epochs=SUPCON_EPOCHS, lr=1e-4, batch_size=BATCH_SIZE)
print("SupCon pretraining finished. Best model saved to:", os.path.join(OUT_DIR, "supcon_best.pth"))
```

```python
# Extract embeddings & compute class prototypes
def extract_embeddings(model, dataset, batch_size=32):
    model = model.to(DEVICE)
    model.eval()
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=False, num_workers=NUM_WORKERS)
    all_embeddings = []
    all_labels = []
    with torch.no_grad():
        for imgs, labels in tqdm(loader, desc="Extract embeddings", leave=False):
            imgs = imgs.to(DEVICE)
            z, feats = model(imgs)
            # use features (feats) or normalized projection z. For prototypes we use feats (higher-d)
            feats = feats
            if isinstance(feats, torch.Tensor) and feats.ndim == 4:
                feats = nn.AdaptiveAvgPool2d((1,1))(feats)
                feats = torch.flatten(feats, 1)
            all_embeddings.append(feats.cpu().numpy())
            all_labels.append(labels.numpy())
    X = np.vstack(all_embeddings)
    y = np.hstack(all_labels)
    return X, y

# Load best SupCon weights
supcon_model.load_state_dict(torch.load(os.path.join(OUT_DIR, "supcon_best.pth"), map_location=DEVICE))
```

```python
# Combine for prototypes
emb_all = np.vstack([emb_train, emb_val])
lbl_all = np.hstack([lbl_train, lbl_val])

# compute class prototypes (mean embedding per class)
prototypes = np.zeros((NUM_CLASSES, emb_all.shape[1]), dtype=np.float32)
for c in range(NUM_CLASSES):
    pts = emb_all[lbl_all == c]
    if len(pts) == 0:
        prototypes[c] = np.zeros(emb_all.shape[1])
    else:
        prototypes[c] = pts.mean(axis=0)

print("Computed class prototypes shape:", prototypes.shape)
```

```python
 #Automatic clustering using silhouette score to pick best #clusters
def pick_best_clusters(prototypes, k_range=(2,5)):
    best_k = None
    best_score = -1.0
    best_labels = None
    for k in range(k_range[0], k_range[1]+1):
        # cluster prototypes (small count)
        clustering = AgglomerativeClustering(n_clusters=k)
        labels = clustering.fit_predict(prototypes)
        if k == 1 or len(set(labels)) == 1:
            continue
        score = silhouette_score(prototypes, labels)
        print(f"k={k}, silhouette={score:.4f}")
        if score > best_score:
            best_score = score
            best_k = k
            best_labels = labels
    return best_k, best_labels, best_score
```

```python
print("Picking best number of groups using silhouette score...")
best_k, best_labels, best_score = pick_best_clusters(prototypes, k_range=(2,5))
```

```python
print("Best k:", best_k, "score:", best_score)
group_labels = best_labels  # array of length NUM_CLASSES mapping class->group
```

```python
# Build Multi-head Joint Model (shared backbone -> embedding -> group head + class head)

class MultiHeadModel(nn.Module):
    def __init__(self, backbone_name='efficientnet_b0', emb_dim=512, num_groups=NUM_GROUPS, num_classes=NUM_CLASSES, pretrained=True):
        super().__init__()
        # backbone
        weights = EfficientNet_B0_Weights.IMAGENET1K_V1 if pretrained else None
        self.backbone = efficientnet_b0(weights=weights)
        # remove original classifier
        in_feats = 1280  # EfficientNet-B0 typical
        self.backbone.classifier = nn.Identity()
        # embedding head
        self.embed = nn.Sequential(
            nn.Linear(in_feats, emb_dim),
            nn.BatchNorm1d(emb_dim),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
        )
        # heads
        self.group_head = nn.Linear(emb_dim, num_groups)
        self.class_head = nn.Linear(emb_dim, num_classes)

    def forward(self, x):
        f = self.backbone(x)
        if f.ndim == 4:
            f = nn.AdaptiveAvgPool2d((1,1))(f)
            f = torch.flatten(f, 1)
        e = self.embed(f)
        g = self.group_head(e)
        c = self.class_head(e)
        return g, c, e

# instantiate
joint_model = MultiHeadModel(emb_dim=512, num_groups=NUM_GROUPS, num_classes=NUM_CLASSES, pretrained=True).to(DEVICE)
print("Joint model created. NUM_GROUPS:", NUM_GROUPS, "NUM_CLASSES:", NUM_CLASSES)
```

```python
#  Joint training (group + class heads) with SupCon-pretrained backbone weights
# We can optionally initialize backbone weights from the SupCon backbone (feature extractor) for better starting point.

def copy_supcon_to_joint(supcon_model, joint_model):
    # supcon_model.backbone is EfficientNet-B0 with classifier removed
    sdict = supcon_model.backbone.state_dict()
    jdict = joint_model.backbone.state_dict()
    # copy matching keys
    for k in jdict.keys():
        if k in sdict and sdict[k].shape == jdict[k].shape:
            jdict[k] = sdict[k]
    joint_model.backbone.load_state_dict(jdict)
    print("Copied supcon backbone weights into joint model backbone (where shapes matched).")

copy_supcon_to_joint(supcon_model, joint_model)
```

```python
# training utilities
def train_joint(model, train_dataset, val_dataset, epochs=10, lr=1e-4, batch_size=32, alpha=0.5, patience=6, out_dir=OUT_DIR):
    # alpha * group_loss + class_loss (alpha ~ 0.5)
    model = model.to(DEVICE)
    # dataloaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=NUM_WORKERS)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=NUM_WORKERS)

    # criterion and optimizer
    ce = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-5)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)

    best_metric = -1.0
    best_wts = copy.deepcopy(model.state_dict())
    no_improve = 0
    history = {'train_loss':[], 'val_loss':[], 'train_acc':[], 'val_acc':[], 'val_macro_f1':[]}

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        running_correct = 0
        running_total = 0
        for imgs, labels in tqdm(train_loader, desc=f"Joint train epoch {epoch+1}/{epochs}", leave=False):
            imgs = imgs.to(DEVICE)
            labels = labels.to(DEVICE)
            # map labels to group labels
            group_targets = torch.tensor([int(group_labels[int(l)]) for l in labels]).to(DEVICE)

            g_logits, c_logits, _ = model(imgs)
            loss_g = ce(g_logits, group_targets)
            loss_c = ce(c_logits, labels)
            loss = alpha * loss_g + (1.0 - alpha) * loss_c  # weight both heads (we can weight class more)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * imgs.size(0)
            # compute accuracy on class head for training monitoring
            _, preds = torch.max(c_logits, dim=1)
            running_correct += (preds == labels).sum().item()
            running_total += imgs.size(0)

        train_loss = running_loss / running_total
        train_acc = running_correct / running_total

        # validation
        model.eval()
        v_running_loss = 0.0
        v_running_correct = 0
        v_total = 0
        all_preds = []
        all_labels = []
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs = imgs.to(DEVICE)
                labels = labels.to(DEVICE)
                group_targets = torch.tensor([int(group_labels[int(l)]) for l in labels]).to(DEVICE)
                g_logits, c_logits, _ = model(imgs)
                loss_g = ce(g_logits, group_targets)
                loss_c = ce(c_logits, labels)
                loss = alpha * loss_g + (1.0 - alpha) * loss_c
                v_running_loss += loss.item() * imgs.size(0)

                _, preds = torch.max(c_logits, dim=1)
                v_running_correct += (preds == labels).sum().item()
                v_total += imgs.size(0)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        val_loss = v_running_loss / v_total
        val_acc = v_running_correct / v_total
        # macro-f1
        try:
            crep = classification_report(all_labels, all_preds, output_dict=True, zero_division=0)
            macro_f1 = crep['macro avg']['f1-score']
        except Exception:
            macro_f1 = 0.0

        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        history['val_macro_f1'].append(macro_f1)

        print(f"[Joint] Epoch {epoch+1}/{epochs} train_loss={train_loss:.4f} val_loss={val_loss:.4f} train_acc={train_acc:.4f} val_acc={val_acc:.4f} macro_f1={macro_f1:.4f}")

        # checkpoint if improved
        if macro_f1 > best_metric + 1e-6:
            best_metric = macro_f1
            best_wts = copy.deepcopy(model.state_dict())
            torch.save(best_wts, os.path.join(out_dir, f"joint_best_epoch{epoch+1}_f1_{best_metric:.4f}.pth"))
            no_improve = 0
        else:
            no_improve += 1

        if no_improve >= patience:
            print("Early stopping (joint) at epoch", epoch+1)
            break

        scheduler.step()

    model.load_state_dict(best_wts)
    return model, history, best_metric
```

```python
# Plot training curves helper
def plot_history(history, title_prefix, save_name):
    epochs = range(1, len(history['train_loss'])+1)
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(epochs, history['train_loss'], label='train_loss')
    plt.plot(epochs, history['val_loss'], label='val_loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f"{title_prefix} Loss")
    plt.legend()
    plt.grid(True)

    plt.subplot(1,2,2)
    plt.plot(epochs, history['train_acc'], label='train_acc')
    plt.plot(epochs, history['val_acc'], label='val_acc')
    plt.xlabel('Epoch')
    plt.ylabel('Acc')
    plt.title(f"{title_prefix} Accuracy")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(OUT_DIR, save_name))
    plt.show()
```

```python
# 11) Specialist fine-tuning per group
# -------------------------
# For each group, filter dataset to that group's classes, remap labels to 0..k-1, instantiating a specialist model starting from joint backbone
def build_group_subset(dataset_paths_list, indices, target_group_label):
    # dataset_paths_list: all_imgs (path,label)
    sel_indices = [i for i in indices if group_labels[all_imgs[i][1]] == target_group_label]
    return PathIndexDataset(all_imgs, sel_indices, train_transforms), sel_indices

# Build path index dataset indices for train/val separately
train_indices_all = train_indices
val_indices_all = val_indices

specialist_models = {}
for g in sorted(groups.keys()):
    # prepare datasets
    # get indices for train subset belonging to group g
    train_idx_g = [i for i in train_indices_all if group_labels[all_imgs[i][1]] == g]
    val_idx_g = [i for i in val_indices_all if group_labels[all_imgs[i][1]] == g]

    if len(train_idx_g) == 0:
        print("Group", g, "has no training samples! skipping.")
        continue

    ds_train_g = PathIndexDataset(all_imgs, train_idx_g, train_transforms)
    ds_val_g = PathIndexDataset(all_imgs, val_idx_g, eval_transforms)

    # remap labels to 0..K-1
    classes_in_g = sorted([i for i in range(NUM_CLASSES) if group_labels[i] == g])
    mapping = {orig: new for new, orig in enumerate(classes_in_g)}
    # wrapper to remap labels
    class RemapDataset(Dataset):
        def __init__(self, base_ds, mapping):
            self.base_ds = base_ds
            self.mapping = mapping
        def __len__(self):
            return len(self.base_ds)
        def __getitem__(self, idx):
            img, lbl = self.base_ds[idx]
            return img, self.mapping[int(lbl)]

    ds_train_g_mapped = RemapDataset(ds_train_g, mapping)
    ds_val_g_mapped = RemapDataset(ds_val_g, mapping)
    num_cls_g = len(classes_in_g)
    print(f"Group {g} -> classes {classes_in_g} ({num_cls_g} classes). Train samples: {len(ds_train_g_mapped)}, Val: {len(ds_val_g_mapped)}")

    # build specialist model: reuse backbone + embed from joint_model
    spec_model = MultiHeadModel(emb_dim=512, num_groups=NUM_GROUPS, num_classes=NUM_CLASSES, pretrained=True)
    # copy joint backbone and embedding weights
    spec_model.backbone.load_state_dict(joint_model.backbone.state_dict(), strict=False)
    spec_model.embed.load_state_dict(joint_model.embed.state_dict(), strict=False)
    # replace class_head to be small specialist head
    spec_model.class_head = nn.Linear(512, num_cls_g)
    spec_model.to(DEVICE)

    # train only head + last layers optionally
    # freeze backbone except embed and class head
    for name, p in spec_model.named_parameters():
        if 'embed' in name or 'class_head' in name:
            p.requires_grad = True
        else:
            p.requires_grad = False

    # optimizer
    optimizer = optim.AdamW(filter(lambda p: p.requires_grad, spec_model.parameters()), lr=1e-4, weight_decay=1e-5)
    ce = nn.CrossEntropyLoss()
    best_f1 = -1.0
    best_w = copy.deepcopy(spec_model.state_dict())
    no_improve = 0
    history_spec = {'train_loss':[], 'val_loss':[], 'train_acc':[], 'val_acc':[]}

    train_loader_g = DataLoader(ds_train_g_mapped, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS)
    val_loader_g = DataLoader(ds_val_g_mapped, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS)

    for epoch in range(SPECIALIST_EPOCHS):
        spec_model.train()
        rl, rc, rt = 0.0, 0, 0
        for imgs, labels in train_loader_g:
            imgs = imgs.to(DEVICE)
            labels = labels.to(DEVICE)
            _, logits, _ = spec_model(imgs)
            loss = ce(logits[:, :num_cls_g], labels)  # class_head replaced so logits shape fits
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            rl += loss.item() * imgs.size(0)
            _, preds = logits.max(1)
            rc += (preds == labels).sum().item()
            rt += imgs.size(0)
        train_loss = rl / rt
        train_acc = rc / rt

        # val
        spec_model.eval()
        vrl, vrc, vrt = 0.0, 0, 0
        all_preds, all_labels = [], []
        with torch.no_grad():
            for imgs, labels in val_loader_g:
                imgs = imgs.to(DEVICE)
                labels = labels.to(DEVICE)
                _, logits, _ = spec_model(imgs)
                loss = ce(logits[:, :num_cls_g], labels)
                vrl += loss.item() * imgs.size(0)
                _, preds = logits.max(1)
                vrc += (preds == labels).sum().item()
                vrt += imgs.size(0)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
        val_loss = vrl / max(1, vrt)
        val_acc = vrc / max(1, vrt)
        try:
            rep = classification_report(all_labels, all_preds, output_dict=True, zero_division=0)
            f1 = rep['macro avg']['f1-score']
        except Exception:
            f1 = 0.0

        history_spec['train_loss'].append(train_loss)
        history_spec['val_loss'].append(val_loss)
        history_spec['train_acc'].append(train_acc)
        history_spec['val_acc'].append(val_acc)

        print(f"[Spec G{g}] Epoch {epoch+1}/{SPECIALIST_EPOCHS} train_loss={train_loss:.4f} val_loss={val_loss:.4f} val_acc={val_acc:.4f} f1={f1:.4f}")

        if f1 > best_f1 + 1e-6:
            best_f1 = f1
            best_w = copy.deepcopy(spec_model.state_dict())
            no_improve = 0
            torch.save(best_w, os.path.join(OUT_DIR, f"spec_group{g}_best_f1_{best_f1:.4f}.pth"))
        else:
            no_improve += 1
        if no_improve >= 4:
            print("Early stopping specialist group", g, "at epoch", epoch+1)
            break

    spec_model.load_state_dict(best_w)
    specialist_models[g] = {
        "model": spec_model,
        "classes": classes_in_g,
        "mapping": mapping,
        "history": history_spec
    }
```

```python
    # save final specialist model
    torch.save(spec_model.state_dict(), os.path.join(OUT_DIR, f"spec_group{g}_final.pth"))

    # plot history for this specialist
    plot_history(history_spec, f"Specialist Group {g}", f"spec_group{g}_history.png")
```

```python
# 12) Final hierarchical inference & evaluation on test set
# -------------------------
def hierarchical_predict(img_tensor, joint_model, specialist_models, tta=False):
    joint_model.eval()
    with torch.no_grad():
        x = img_tensor.unsqueeze(0).to(DEVICE)
        g_logits, c_logits, _ = joint_model(x)
        g_pred = int(torch.argmax(g_logits, dim=1).item())
        # use specialist for group g if available
        if g_pred in specialist_models:
            spec = specialist_models[g_pred]["model"]
            spec.eval()
            with torch.no_grad():
                _, s_logits, _ = spec(x)
                # s_logits maps to indices 0..k-1
                s_pred = int(torch.argmax(s_logits, dim=1).item())
                # map back to original class
                orig_class = specialist_models[g_pred]["classes"][s_pred]
                return orig_class
        else:
            # fallback to joint class head
            return int(torch.argmax(c_logits, dim=1).item())
```

```python
# Run evaluation on test set
y_true = []
y_pred = []
for imgs, labels in tqdm(test_loader, desc="Final test inference"):
    for i in range(imgs.size(0)):
        img = imgs[i]
        true = int(labels[i].item())
        pred = hierarchical_predict(img, joint_model, specialist_models)
        y_true.append(true)
        y_pred.append(pred)

print("\nFinal Classification Report (hierarchical):")
print(classification_report(y_true, y_pred, target_names=class_names, zero_division=0))
```
